```python
## Reproduce linear regression
def compute_nonskewed_samples_linreg(n, samples=100):
    X = np.random.normal(size=(n,d))
    w = np.random.normal(size=(d))

    y = X @ w
    w_pred = np.linalg.lstsq(X,y)[0]
    
    new_x = np.random.normal(size=(samples,d))
    return np.mean((new_x @ w - new_x @ w_pred)**2/d)
```

```python
def trials_nonskewed_compute_samples_linreg(n,trials=1000):
    s = 0
    for _ in range(trials):
        s += compute_nonskewed_samples_linreg(n)
    return s/trials
```

```python
def generate_random_cov(d):
    import scipy.linalg as la
    des = np.random.rand(d) * np.array([1/i**2 for i in range(1,d+1)])

    s = np.diag(des)
    q, _ = la.qr(np.random.rand(d, d))
    semidef = q.T @ s @ q
    return semidef
```

```python
## Reproduce linear regression
def compute_skewed_samples_linreg(n, samples=1000):

    cov = generate_random_cov(d)
    X = np.random.multivariate_normal(np.zeros(d), cov, size=(n))
    w = np.random.normal(size=(d))

    y = X @ w
    w_pred = np.linalg.lstsq(X,y)[0]
    
    new_x = np.random.multivariate_normal(np.zeros(d), cov, size=(samples))
    print(len((new_x @ w - new_x @ w_pred)**2/d))
    return np.mean((new_x @ w - new_x @ w_pred)**2/d)

def trials_skewed_compute_samples_linreg(n,trials=10):
    s = 0
    for _ in range(trials):
        s += compute_skewed_samples_linreg(n)
    return s/trials
```
